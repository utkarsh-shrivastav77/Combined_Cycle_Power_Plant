# Gradient Descent
Applying the gradient Descent Algorithm  using the sklearn library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error

from sklearn import preprocessing
```

## Loading the dataset

```python
data = pd.read_csv('Test_data.csv')
df = pd.DataFrame(data)
df.columns.values[4] = 'EP'
df
```

```python
df.describe()
```

## Spliting the data 
This is done because this data has both features and target values.So, we will split the datasets in X_train and Y_train which will contain features and target values respectively

```python
X_train = df.iloc[:,:4]
Y_train = df.iloc[:,-1]
Y_train
```

## Applying the Gradient descent
Using the sklearn library of linear regression to train the data and predict the outcome

```python
model = GradientBoostingRegressor()
model.fit(X_train,Y_train)
```

## Loading the Test Datasets

```python
test = pd.read_csv('test_data.csv')
td = pd.DataFrame(test)
td.columns.values[4] = 'EP'
final_train = td.iloc[:,:4]
final_test = td.iloc[:,-1]
final_test
```

## Prediction of the dataset

```python
y_predict = model.predict(final_train)
y_prediction = pd.DataFrame(y_predict)
y_prediction
```

```python
y_prediction.columns = ["EP_Predicted"]
y_prediction
```

```python
print(mean_absolute_error(final_test,y_predict))
```

```python
combine = [final_test,y_prediction]
final_data = pd.concat(combine,axis=1)
final_data
```

```python
final_data.to_csv('data.csv')
```
